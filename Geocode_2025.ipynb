```python
# Step 1: Install required libraries
!pip install pandas openpyxl

# Step 2: Import libraries
import pandas as pd
import requests
from google.colab import files
import math # Import math for checking for NaN

# Step 3: Upload the Excel file
uploaded = files.upload()

# Step 4: Load the Excel file
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

# Check for 'Address' column
if 'Address' not in df.columns:
    raise ValueError("The uploaded Excel file must contain a column named 'Address'.")

# Step 5: Define the Google Maps Geocoding function
# Replace with your actual API key, the current one is just an example.
API_KEY = "AIzaSyBWx4p2-kmBlpbh2eECzQ6VJr3xS8wOrA0"

def geocode_address(address):
    # Ensure the address is a string and not empty or NaN
    if not isinstance(address, str) or not address.strip() or pd.isna(address):
        return None, None # Return None for invalid addresses

    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={requests.utils.quote(address)}&key={API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        results = response.json().get("results")
        if results:
            location = results[0]["geometry"]["location"]
            return location["lat"], location["lng"]
    return None, None

# Step 6: Apply geocoding
latitudes = []
longitudes = []

for address in df['Address']:
    # Pass the address to the function which now handles invalid types
    lat, lng = geocode_address(address)
    latitudes.append(lat)
    longitudes.append(lng)

df["Latitude"] = latitudes
df["Longitude"] = longitudes

# Step 7: Save to new Excel file
output_file = "geocoded_addresses.xlsx"
df.to_excel(output_file, index=False)

# Step 8: Download the file
files.download(output_file)
```
